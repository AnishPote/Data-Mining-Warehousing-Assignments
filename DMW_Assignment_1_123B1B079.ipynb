```python
df = df.drop_duplicates(subset=['Name', 'Age', 'Department', 'Salary', 'Experience_Years']).reset_index(drop=True)
df
```

```python
df['Age'].fillna(df['Age'].mean())
df['Salary'].fillna(df['Salary'].mean())
df['Experience_Years'].fillna(df['Experience_Years'].mean())
df
```

```python
df_minmax = df.copy()
for col in ['Age', 'Salary', 'Experience_Years']:
    df_minmax[col] = (df_minmax[col] - df_minmax[col].min()) / (df_minmax[col].max() - df_minmax[col].min())
df_minmax
```

```python
print("\n Central Tendency:")
print("Mean:\n", df[['Age', 'Salary', 'Experience_Years']].mean())
print("Median:\n", df[['Age', 'Salary', 'Experience_Years']].median())
print("Mode:\n", df[['Age', 'Salary', 'Experience_Years']].mode().iloc[0])
```

```python
print("\n📊 Dispersion:")
print("Variance:\n", df[['Age', 'Salary', 'Experience_Years']].var())
print("Standard Deviation:\n", df[['Age', 'Salary', 'Experience_Years']].std())
print("Range:\n", df[['Age', 'Salary', 'Experience_Years']].max() - df[['Age', 'Salary', 'Experience_Years']].min())
```
